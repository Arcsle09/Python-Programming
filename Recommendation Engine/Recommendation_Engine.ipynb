```python
##Step 1: Read CSV File
movies = pd.read_csv('movie_dataset.csv')
print(movies.columns)
```

```python
##Step 2: Select Features
movies_df = movies
features = ['keywords','cast','genres','director']
for i in features:
    movies_df[i] = movies_df[i].fillna('')
movies_df.describe()
```

```python
##Step 3: Create a column in DF which combines all selected features
def combine_features(row):
    try:
        return row['keywords'] + " " + row['cast'] + " " + row['genres'] + " " + row['director']
    except:
        print("Error: ",row)
movies_df['combined_features'] = movies.apply(combine_features,axis=1)
print(movies_df['combined_features'].head())
```

```python
##Step 4: Create count matrix from this new combined column
cv = CountVectorizer()
type(cv)
count_matrix=cv.fit_transform(movies_df['combined_features'])
```

```python
#Step5: compute the cosine similarity based on the count_matrix
cosine_sim = cosine_similarity(count_matrix)
print(cosine_sim)
```

```python
###### helper functions. Use them when needed #######
def get_title_from_index(index):
	return movies_df[movies_df.index == index]["title"].values[0]

def get_index_from_title(title):
	return movies_df[movies_df.title == title]["index"].values[0]
##################################################
```

```python
movie_user_likes = 'Avatar'
movies_index = get_index_from_title(movie_user_likes)
```

```python
movies_index
```

```python
similar_movies = list(enumerate(cosine_sim[movies_index]))
sorted_similar_movies = sorted(similar_movies,key=lambda x:x[1],reverse = True)
```
